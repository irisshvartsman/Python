```python
headers= ('id', 'date', 'tweet', 'fave', 'retweet')
file = pd.read_csv('sampletwitter.csv', header=0)
file.columns = headers
file.head(10)
```

```python
import re
def data_cleansing(corpus):
        remove_hash = re.sub("#\S+", " ", str(corpus))

        remove_pic = re.sub("pic\S+", " ", str(remove_hash))

        remove_at = re.sub("@\S+", " ", str(remove_pic))

        remove_special = re.sub("[^a-zA-Z.\d\s]", " ", str(remove_at))

        remove_rt = re.sub("^b.RT", " ", str(remove_special))

        words = remove_rt.lower().split()
        return (" ".join(words))
```

```python
file['tweet'] = file['tweet'].apply(lambda x:data_cleansing(x))
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
def sentiment_val(text):
    analyser = SentimentIntensityAnalyzer()
    result = analyser.polarity_scores(text)
    score = result['compound']
    return round(score,1)
```

```python
#running a sample on one line just make sure the sentiment analyser works

sample = file['tweet'][20]
print(sample)
print(sentiment_val(sample))
```

```python
tweets = file['tweet']
all_sent_val = []
all_sentiments=[]

for i in range(0, len(file)):
    all_sent_val.append(sentiment_val(tweets[i]))
```
